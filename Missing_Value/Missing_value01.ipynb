Impputation Technique

1-Mean Value Imputation

```python
sns.distplot(df['age'])
```

This graph is normal distribution 

In Normal distribution we replace NAN value with mean 

```python
df['Age_mean'] = df['age'].fillna(df['age'].mean())
```

```python
df[['Age_mean', 'age']]
```

This technique work will when your data normal distribution

***Median Value Implementation***

if you have outliers in data use this technique

```python
df['Age_median'] = df['age'].fillna(df['age'].median())
```

```python
df[['Age_median','Age_mean', 'age']]
```

***Mode Value Computation***

Mostly used in categorical value

```python
df[df['embarked'].isnull()]
```

```python
df['embarked'].unique()
```

```python
df['age'].notna() #false means there valaue is NAN
```

```python
df[df['age'].notna()]
```

```python
df['embarked'].mode()[0]
```

```python
df[df['age'].notna()]['embarked'].mode()[0]
```

```python
mode = df[df['age'].notna()]['embarked'].mode()[0]
```

```python
df['embarked_mode'] = df['embarked'].fillna(mode)
```

```python
df[['embarked_mode', 'embarked']]
```

```python
df['embarked_mode'].isnull().sum()
```
