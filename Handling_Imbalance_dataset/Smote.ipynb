```python
from sklearn.datasets import make_classification
# for making classified data set
```

```python
# x independent features
# y is dependent features (output)
x,y  =make_classification(n_samples=1000, n_features=2,n_redundant=0, n_clusters_per_class=1, weights=[0.90], random_state=1)
# n_cluster_per_class = cluster means group and one class have one cluster
# weight means make data balance or imbalance 0.90 is imbalance
# random state any value or 1
# redundant study in ML 
```

```python
x,y
```

Convert in DataFrame

```python
import pandas as pd
df1 = pd.DataFrame(x, columns=['f1', 'f2'])
df2 = pd.DataFrame(y, columns=['target'])
final_df = pd.concat([df1,df2],axis=1) #use axis and axis =1 because concat the data according to columna
```

```python
final_df.head()
```

```python
final_df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['f1'], final_df['f2'], c = final_df['target'])
```

```python
# treansform the dataset
oversample =  SMOTE()
x, y = oversample.fit_resample(final_df[['f1', 'f2']], final_df['target'])
```
